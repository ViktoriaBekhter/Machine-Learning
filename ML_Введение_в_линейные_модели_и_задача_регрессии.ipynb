```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Обучающие данные
X_train = np.array([[1], [2], [3], [4], [5]])
y_train = np.array([2, 4, 5, 4, 5])

# Создание и обучение модели
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Предсказание на обучающих данных
y_pred_train = regressor.predict(X_train)

# Расчет метрик
mae = mean_absolute_error(y_train, y_pred_train)
mse = mean_squared_error(y_train, y_pred_train)
rmse = np.sqrt(mse)
r2 = r2_score(y_train, y_pred_train)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R^2 score:", r2)

# Построение графика остатков
residuals = y_train - y_pred_train
plt.scatter(y_pred_train, residuals, color='green')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
#12. Оценить линейную регрессию методом максимального правдоподобия при помощи функции neg_log_likelihood.
import numpy as np
from scipy.optimize import minimize

# Определение функции отрицательного логарифма правдоподобия
def neg_log_likelihood(params, x, y):
    intercept, slope = params
    y_pred = intercept + slope * x
    neg_ll = -np.sum(np.log(np.exp(-(y - y_pred)**2 / 2) / np.sqrt(2*np.pi)))
    return neg_ll

# Данные для обучения
x_train = np.array([1, 2, 3, 4, 5])
y_train = np.array([2, 4, 6, 8, 10])

# Начальные значения параметров
initial_params = [0, 0]

# Оптимизация параметров с помощью метода Nelder-Mead
result = minimize(neg_log_likelihood, initial_params, args=(x_train, y_train), method='Nelder-Mead')

# Получение оптимальных параметров
intercept_ml, slope_ml = result.x

print("Оценка методом максимального правдоподобия:")
print("Перехват (intercept):", intercept_ml)
print("Наклон (slope):", slope_ml)
```

```python
from scipy.optimize import minimize

# Данные
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 5, 4, 5])

# Начальные значения для оптимизации
initial_guess = [0, 0]

# Минимизация отрицательного логарифма правдоподобия
result = minimize(neg_log_likelihood, initial_guess, args=(x, y))

# Оценки коэффициентов
intercept_ml, slope_ml = result.x

print("Оценка intercept (точка пересечения с осью ординат):", intercept_ml)
print("Оценка slope (наклон):", slope_ml)
```

```python
# Шаг 14: Прогнозные значения, среднеквадратичная ошибка и коэффициент детерминации для линейной регрессии
y_pred = slope_ml * x + intercept_ml
mse_ml = mean_squared_error(y, y_pred)
r2_ml = r2_score(y, y_pred)

print(f'Прогнозные значения для линейной регрессии: {y_pred}')
print(f'Среднеквадратичная ошибка для линейной регрессии: {mse_ml}')
print(f'Коэффициент детерминации для линейной регрессии: {r2_ml}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr

# Пример данных
X = np.array([[1], [2], [3], [4], [5]])  # Признаки
y = np.array([2, 3, 4, 5, 6])  # Целевая переменная
k = 3  # Количество соседей

# Обучение модели K ближайших соседей
knn_model = KNeighborsRegressor(n_neighbors=k)
knn_model.fit(X, y)

# Прогнозирование
y_pred = knn_model.predict(X)

# Расчет метрик качества модели
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
me = np.mean(y - y_pred)
pearson_corr = pearsonr(y, y_pred)[0]
r2 = r2_score(y, y_pred)
residuals = y - y_pred

# Построение графика остатков
plt.scatter(y_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Error:", me)
print("Pearson Correlation Coefficient:", pearson_corr)
print("R-squared:", r2)
print("Residuals:", residuals)
```
