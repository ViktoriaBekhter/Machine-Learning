10. Рассчитать проекции объектов на главные компоненты.

```python
import pandas as pd
from sklearn.decomposition import PCA

data = pd.read_csv('pima-indians-diabetes.data.csv')
pca = PCA(n_components=2)
pca.fit(data)
principal_components = pca.transform(data)
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

print(principal_df)
```

13. Выведите на экран значения дисперсий по компонентам и общую объясненную дисперсию. Выведите график объясненных дисперсий по компонентам.

```python
explained_variance_ratio = pca.explained_variance_ratio_
total_explained_variance = np.sum(explained_variance_ratio)

print("Дисперсия по компонентам:", explained_variance_ratio)
print("Общая объясненная дисперсия:", total_explained_variance)

plt.bar(range(1, len(explained_variance_ratio)+1), explained_variance_ratio)
plt.xlabel('Главные компоненты')
plt.ylabel('Объясненная дисперсия')
plt.show()
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('pima-indians-diabetes.data.csv')

# Создаем ковариационную матрицу
cov_matrix = np.cov(df.T)

print("Ковариационная матрица:")
print(cov_matrix)

# Вычисляем собственные значения и векторы
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\nСобственные значения:")
print(eigenvalues)
```

```python
import matplotlib.pyplot as plt

# Визуализируем собственные значения
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, marker='o')
plt.xlabel('Факторы')
plt.ylabel('Собственные значения')
plt.title('Собственные значения от факторов')
plt.grid(True)
plt.show()
```
